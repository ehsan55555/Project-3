```python
# Convert the 'persons_injured' column to integers
new_york_crash_df['persons_injured'] = new_york_crash_df['persons_injured'].astype(int)

# Replace NaN values in the 'persons_injured' column with -1 (or any other placeholder)
new_york_crash_df['persons_injured'].fillna(-1, inplace=True)

# Convert the 'persons_injured' column to integers
new_york_crash_df['persons_injured'] = new_york_crash_df['persons_injured'].astype(int)
```

```python
def get_borough(lat, lon):
    # Define a dictionary mapping borough names to their latitude and longitude ranges
    boroughs = {
        "MANHATTAN": [(40.7000, 40.8800), (-74.0200, -73.9100)],
        "BRONX": [(40.7900, 40.9200), (-73.9300, -73.7800)],
        "BROOKLYN": [(40.5700, 40.7400), (-74.0400, -73.8700)],
        "QUEENS": [(40.5400, 40.8000), (-74.0300, -73.7000)],
        "STATEN ISLAND": [(40.4800, 40.6500), (-74.2500, -74.0800)]
    }

    # Iterate through the boroughs and check if the coordinates fall within a range
    for borough, (lat_range, lon_range) in boroughs.items():
        if lat_range[0] <= lat <= lat_range[1] and lon_range[0] <= lon <= lon_range[1]:
            return borough

    # If no match is found, return None
    return None
```

```python
new_york_crash_df['borough'] = new_york_crash_df.apply(
    lambda row: get_borough(row['latitude'], row['longitude']),
    axis=1
)

# Fill any remaining missing values in the 'borough' column with "Unknown"
new_york_crash_df['borough'].fillna("Unknown", inplace=True)
```

```python
# Filter rows with missing "borough" values
missing_borough = new_york_crash_df[new_york_crash_df['borough'].isna()].copy()

# Apply the get_borough function to fill in missing values
missing_borough['borough'] = missing_borough.apply(lambda row: get_borough(row['latitude'], row['longitude']), axis=1)

# Update the original DataFrame with filled values using .loc
new_york_crash_df.loc[missing_borough.index, 'borough'] = missing_borough['borough']
```

```python
columns_to_drop = ['lat_and_lon', 'zip_code', 'on_street', 'cross_street', 'off_street']

# Drop the specified columns
new_york_crash_df = new_york_crash_df.drop(columns=columns_to_drop)


new_york_crash_df
```

```python
# Convert 'crash_time' to a datetime object
new_york_crash_df['crash_time'] = pd.to_datetime(new_york_crash_df['crash_time'])

# Format 'crash_time' as a user-friendly string
new_york_crash_df['crash_time'] = new_york_crash_df['crash_time'].dt.strftime('%I:%M %p')

new_york_crash_df
```
